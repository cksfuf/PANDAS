### loc - 조건필터링

```python
# 컬럼 비교
cond = df['who'] == 'man'
cond
# 데이터 프레임 중 who 로 들어가 man과 일치하는 데이터를 Ture 로 반환해주는 시리즈 데이터로 보여줌
```

- case1: df[condition]

```python
cond = df['who'] == 'man'
df[cond].head()
# 전체 데이터 중 남성 데이터만 출력한것
# 데이터 프레임 자체에 넣는 방법
# 근데 우린 이 식 잘 안쓸거임
    # 이유는 값을 직접 수정할때 오류 발생
    # 아래에 코드 실행했음 보셈.
```

- case2: df.loc[condition]

```python
cond = df['who'] == 'man'
df.loc[cond].head()
```

```python
# 왜 첫번째 식을 잘 사용하지 않을거냐?
cond = df['who'] == 'man'
df[cond]['age'] = 100
# 데이터 프레임에 직접 값을 추가하려면 오류가 생김
```

```python
cond = df['who'] == 'man'
df.loc[cond,'age'] = 100
df.head()
# 데이터 중 남자 데이터 age 부분이 전부 100이 들어감
```

```python
# 조건 2개 만들기
cond1 = df['fare'] > 30
cond2 = df['who'] == 'woman'

# 조건 2개 만족하는 식 &
df.loc[cond1 & cond2].head()
```

```python
# 둘 중 하나만 맞아도 되는 경우
df.loc[cond1 | cond2].head()
```

```python
# loc 랑 비슷한 코드
# iloc: index로만 구별
df.iloc[1, 7]
# 많이 사용하지 않으니 알고만 있으셈.
```

### where

### practice

```python
tips = sns.load_dataset('tips')
tips.head()
```

- 남성 흡연자 조건 필터링
- tip을 많이 낸 순서로 내림차순 정렬
- 상위 10개 출력

```python
# 남성 흡연자 조건 필터링
cond1 = tips['sex'] == 'Male'
cond2 = tips['smoker'] == 'Yes'

# tip 내림차순 + 상위 10개
tips.loc[cond1 & cond2].sort_values('tip', ascending=False).head()
```

- 식사인원이 5명 이상인 저녁테이블
- 흡연자가 없는 주말 테이블
- total_bill, tip 출력

```python
cond1 = tips['size'] > 4
cond2 = tips['time'] == 'Dinner'
cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sat', 'Sun'])

# tip 내림차순 + 상위 10개
tips.loc[(cond1 & cond2) & (cond3 & cond4), ['total_bill', 'tip']].head()
```
