### 엑셀 - save

```python
# 엑셀 파일 가져오기
excel = pd.read_excel('data/DMF1.xlsx', sheet_name='info', engine='openpyxl')

# 엑셀 수정하기(강사님 이름 수정)
student = excel[1:]

# 새로운 셀로 저장(data 폴더에 student 이름으로 xlsx 형태 저장)
student.to_excel('data/student.xlsx')
```

## CSV - save

```python
# csv 파일 불러오기
df = pd.read_csv('data/DMF1.csv')

# 파일 수정
student = df[1:]

# 파일 저장
student.to_csv('data/student.csv', index=False)
# student.to_excel('data/student.excel', index=False) -> excel 형식으로도 저장 가능.
```
