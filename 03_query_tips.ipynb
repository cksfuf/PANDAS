```python
df_ti = sns.load_dataset('titanic')
df2 = df_ti.copy()

# 남자 승객의 나이 평균
    # m_age = df2.loc[df2['sex'] == 'male', 'age'].mean()
    # 더 편한 방법
c1 = df2['sex'] == 'male'
m_age = df2.loc[c1, 'age'].mean() # df2['age'].sum() / df2['age'].value_counts().sum() 랑 같음

# 남자 승객 결측치에 평균 나이 넣기
df2 = df2.loc[c1, 'age'].fillna(m_age)
df2
```
