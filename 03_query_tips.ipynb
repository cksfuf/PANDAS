### titanic
- age가 결측치인 사람들의 나이를 30살로 채워주세요.
- 남자 승객의 나이 결측치 -> 남자 승객의 나이 평균
- 여자 승객의 나이 결측치 -> 여자 승객의 나이 평균
  
>모든승객의 나이 평균: 29.736

````python
df_ti = sns.load_dataset('titanic')
df2 = df_ti.copy()
df2.head(8)
````

````python
df2.isna().sum()
````

````python
# age 결측값에 30 넣기
df2['age'].fillna(30)
# 이 값은 저장 안됨
````

````python
df_ti = sns.load_dataset('titanic')
df2 = df_ti.copy()

# 남자 승객의 나이 평균
    # m_age = df2.loc[df2['sex'] == 'male', 'age'].mean()
    # 더 편한 방법
c1 = df2['sex'] == 'male'
m_age = df2.loc[c1, 'age'].mean()

# 남자 승객 결측치에 평균 나이 넣기
# df2 = df2.loc[c1, 'age'].fillna(m_age)
    # 이렇게 하면 df2 전체가 바뀜 정확히 c1, 'age' 이부분에 할당해야함.
df2.loc[c1, 'age'] = df2.loc[c1, 'age'].fillna(m_age)
df2.head()
````

````python
# 여자 승객의 나이 평균
c2 = df2['sex'] == 'female'
fm_age = df2.loc[c2, 'age'].mean()

# 여자 승객 결측치에 평균 나이 넣기
df2.loc[c2, 'age'] = df2.loc[c2, 'age'].fillna(fm_age)
df2.head()
````

````python
df2['age'].mean()
````
