### fillna
- 결측값에 특정 값을 채울때 씀

```python
df = sns.load_dataset('titanic')
```

```python
df1 = df.copy()
```

```python
# 결측값에 100을 채움
df1['age'].fillna(100).tail()
```

```python
df1.tail()
# fillna는 결과만 보여주고 저장이 안됨
```

```python
df1['age'] = df1['age'].fillna(100)
```

```python
df1.tail()
```

```python
# 문자열 .fillna()
df1['deck'].fillna('C')
```

```python
# Categories (7, object): ['A', 'B', 'C', 'D', 'E', 'F', 'G'] 범위에 없는 값을 추가하면?
# df['decf'].fillna('No data')
# 에러가 뜸
```

```python
# 카테고리 범주에 없는 없는 값 넣기
df1['deck'] = df1['deck'].cat.add_categories('No data')
df1['deck'].fillna('No data')
```

```python
df1 = df.copy()
```

```python
mean_age = df1['age'].mean()
df1['age'].fillna(mean_age).tail()
```

```python
median_age = df1['age'].median()
df1['age'].fillna(median_age).tail()
```

- 남자 승객의 결측치 -> 남자 승객의 평균
- 여자 승객의 결측치 -> 여자 승객의 평균

 -> 모든승객의 나이 평균: 29.736

```python
df1 = df.copy()
df1.head()
```

```python
# 남자승색의 결측치 채우기
c1 = df1['sex'] == 'male'
male_mean = df1.loc[c1, 'age'].mean()
df1.loc[c1, 'age'] = df1.loc[c1, 'age'].fillna(male_mean)

# 여자승객의 결측치 채우기
c2 = df1['sex'] == 'female'
female_mean = df1.loc[c2, 'age'].mean()
df1.loc[c2, 'age'] = df1.loc[c2, 'age'].fillna(female_mean)

df1['age'].mean()
```

```python
df1['deck'].fillna(df1['deck'].mode()[0]).head()
```
