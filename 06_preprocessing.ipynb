### to_numeric
- 열을 숫자로 변환합니다

```python
df = pd.read_csv('data/seoul_bicycle.csv')
df1 = df.copy()
df1.head()
# 운동량 데이터 확인
```

```python
df1.info()
# 운동량 또한 object 로 되어있네
```

```python
pd.to_numeric(df1['운동량'])
# 2344 데이터에서 오류가 생김
# 확인해보자
```

```python
df1.loc[2344]
# 데이터 결측 확인
```

```python
df1['운동량'] = pd.to_numeric(df1['운동량'], errors='coerce')
df1.info()
# pd.to_numeric(): 열을 숫자로 변환합니다.
# 'errors='coerce':변환 과정에서 오류가 발생한 경우, 해당 값을 NaN(Not a Number)으로 설정. 오류가 발생해도 코드가 멈추지 않고 NaN으로 대체.
```

```python
df1.loc[2344]
```

```python
# 운동량 데이터 분석
df1.tail()
```

```python
bins = [0, 1000, 2000, df1['운동량'].max()]
pd.cut(df1['운동량'], bins).value_counts()
# pd.cut(df1['운동량'], bins): Pandas의 cut 함수를 사용하여 '운동량' 열의 각 값을 bins에 정의된 구간에 맞게 분할
```

```python
pd.cut(df1['운동량'], bins, labels=['부족', '보통', '많음'])
```

```python
df1.describe()
```

```python
# 운동량을 값이 아닌 분위로 나누어 확인
df1['운동량_cut'] = pd.cut(df1['운동량'], bins=10)
df1['운동량_cut'].value_counts()
# 범위를 값을 기준으로 나눔
# 첫 구간에 데이터가 몰려있는 것을 확인
# 뒷 구간에 데이터가 없는것을 확인
```

```python
df1['운동량_qcut'] = pd.qcut(df1['운동량'], q=10)
df1['운동량_qcut'].value_counts()
# 구간을 보면
# (3328.186, 6805.188]
# (6805.188, 163936052.3])
# 에서 갑자기 커지는 것을 확인해볼 수 있음

# 여기에 이상치가 있음을 확일할 수있음.
```

```python
qcut_bins = [0, 0.2, 0.8, 1]
pd.qcut(df1['운동량'], qcut_bins)
```

```python
df1['운동량'] = pd.qcut(df1['운동량'], qcut_bins, labels=['적음', '보통', '많음'])
df1.head()
```
